```python
file = open('/home/lindsey/Documents/BNZ_chat1.txt',mode='r',encoding="utf8")
data = file.read()
#file.close()

regex_time = r'\d{2}:\d{2}:\d{2}'
compiled_regex_time = re.compile(regex_time)
times = re.findall(compiled_regex_time, data)

regex_author = r'\bFrom \s(.*?:)'
compiled_regex_author = re.compile(regex_author)
authors = re.findall(compiled_regex_author, data)

regex_comment = r'\b.*(?=(\\n))'   #r'(?:\: )(.*$)'

# compiled_regex is a Pattern object


#A quick look at who wrote comments and private comments to remove
count_authors={}
for each in authors:
    if each in count_authors.keys():
        count_authors[each]+=1
    else:
        count_authors[each]=1
count_authors
```

```python
messages_split = compiled_regex_author.split(data)
messages_split
```

```python
authors = messages_split[1::2]
print(len(authors))
```

```python
time = []
for each in messages_split[0::2]:
    time.append(re.findall(regex_time, each))

print(time[-1])
time=time[:-1]
print(time[-1])
print(len(time))
```

```python
comments = []
for each in messages_split[0::2]:
    if '\n' in each:
        comments.append(each.split('\n')[0])

comments
```

```python
sep_msgs = []
for each in count_authors.keys():
    for msg in range(len(messages_split)):
        if each == messages_split[msg]:
            sep_msgs.append(messages_split[msg+1]) 
            #getting message after sender along with times

#checking that loop worked
print(len(sep_msgs)) #number of messages
sep_msgs[7] #print message 7
```

```python
cleaned_sep_msg = []

for each in sep_msgs:
    if '\n' in each:
        cleaned_sep_msg.append(each.split('\n'))

        
my_msg = []
for each in cleaned_sep_msg:
    my_msg.append(each[0])

    
print(len(my_msg))
print(my_msg[7])
```

```python
# I noticed a few steps down that the time column is out of order
#reordering the time stamps and paired comments
bnz1_df = pd.DataFrame()
#bnz1_df['time'] = time
#bnz1_df['comment'] = my_msg
bnz1_df = pd.DataFrame(list(zip(time, authors, comments)), 
               columns =['time','author', 'comment']) 
#bnz1_df = bnz1_df.sort_values('time').reset_index(drop=True)
bnz1_df
```

```python
#adding the author column and putting columns in the order I like
bnz1_df['author'] = authors
bnz1_df = bnz1_df[['time','author','comment']]
#remove brackets from time column
bnz1_df['time']= bnz1_df['time'].str.get(0)
#remove private messages
bnz1_df = bnz1_df[~bnz1_df['author'].str.contains("Privately")].reset_index(drop=True)
#remove colon from author column
#this replaces all punctuation. Overkill right now but handy sometimes
bnz1_df['author'] = bnz1_df['author'].str.replace(r'[^\w\s]+', '')



bnz1_df.head()
```

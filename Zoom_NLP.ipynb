```python
file = open('/home/lindsey/Documents/BNZ_chat1.txt',mode='r',encoding="utf8")
data = file.read()
file.close()

#pattern = re.compile('\d+:\d+\s+-\s+([a-zA-Z0-9]+\s?[a-zA-Z0-9]+\s?[a-zA-Z0-9]+\s?):\s+')
#messengers = re.findall(pattern,data)


regex_time = r'\d{2}:\d{2}:\d{2}'
compiled_regex_time = re.compile(regex_time)
times = re.findall(compiled_regex_time, data)

regex_author = r'\bFrom \s(.*?:)'
compiled_regex_author = re.compile(regex_author)
authors = re.findall(compiled_regex_author, data)

regex_comment = r'(?:\: )(.*$)'

# compiled_regex is a Pattern object

count_authors={}
for each in authors:
    if each in count_authors.keys():
        count_authors[each]+=1
    else:
        count_authors[each]=1
count_authors
```

```python
messages_split = compiled_regex_author.split(data)
messages_split
```

```python
authors = messages_split[1::2]
print(len(authors))
authors = authors[:-1]
print(len(authors))
```

```python
sep_msgs = []
for each in count_authors.keys():
    for msg in range(len(messages_split)):
        if each == messages_split[msg]:
            sep_msgs.append(messages_split[msg+1]) 
            #getting message after sender along with times

#checking that loop worked
print(len(sep_msgs)) #number of messages
sep_msgs[7] #print message 7
```

```python
cleaned_sep_msg = []
cleaned_sep_time = []
for each in sep_msgs:
    cleaned_sep_time.append(re.findall(regex_time, each))
    if '\n0' in each:
        cleaned_sep_msg.append(each.split('\n0'))
    elif '\n1' in each:
        cleaned_sep_msg.append(each.split('\n1'))
        
my_msg = []
for each in cleaned_sep_msg:
    my_msg.append(each[0])
    
print(len(my_msg))
print(my_msg[7])
print(cleaned_sep_time[7])
```

```python
bnz1_df = pd.DataFrame(list(zip(cleaned_sep_time, authors, my_msg)), 
               columns =['time', 'author', 'comment']) 
bnz1_df 
```
